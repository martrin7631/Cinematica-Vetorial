```python
# Centro Universitário FEI

# Elaborado por Marcelo Santos Trindade em 18/06/2023

# "Elaboração de programas em Python como apoio didático na disciplina de Física I para estudantes ingressantes".

import matplotlib.pyplot as plt
import numpy as np


# Função para imprimir o texto em verde ou vermelho
def print_verde(texto):
    print('\033[32m{}\033[0m'.format(texto))


def print_vermelho(texto):
    print('\033[31m{}\033[0m'.format(texto))


# Definindo as coordenadas do vetor posição inicial do bloco
x0 = 0
y0 = 2.5

# Definindo a magnitude da força
forca = 100

# Definindo o ângulo da força em graus
angulo = 38

# Convertendo o ângulo para radianos
angulo_rad = np.deg2rad(angulo)

# Calculando as componentes x e y da força
forca_x = forca * np.cos(angulo_rad)
forca_y = forca * np.sin(angulo_rad)

fig, ax = plt.subplots(figsize=(4, 4))
ax.set_aspect('equal')

# Adicionando uma linha horizontal
plt.axhline(y=0, color='black')

# Desenhando o quadrado
quadrado = plt.Rectangle((-2.5, 0), 5, 5, edgecolor='black', facecolor='none')
plt.gca().add_patch(quadrado)

# Plotando o gráfico com o vetor força
plt.quiver(x0, y0, forca_x, forca_y, angles='xy', scale_units='xy', scale=1, color='r')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xticks([])
plt.yticks([])
plt.title('Cinemática vetorial')


print('                                CINEMÁTICA VETORAIL                                ')
print('---------------------------------------------------------------------------------------------')
print('A cinemática vetorial é um ramo da física que estuda o movimento dos objetos usando vetores. \n'
      'Ela descreve a posição, velocidade e aceleração de um objeto usando vetores para representar \n'
      'magnitude, sentido e direção. Isso permite uma análise mais completa do movimento, levando em \nconsideração '
      'diferentes direções e trajetórias.')
print('---------------------------------------------------------------------------------------------')

plt.show()
Fx = round(forca_x, 2)
Fy = round(forca_y, 2)
g = 10
P = 25*g
Normal = P - Fy
a = Fx/25
#print(Fx, Fy)

print('---------------------------------------------------------------------------------------------')
print('                                       QUESTÃO 1                                             ')
print('---------------------------------------------------------------------------------------------')
print('A figura mostra um bloco sob ação de uma força, representada pelo vetor vermelho. \n\n'
      'Dados:\n'
      'massa do bloco = 25kg\n'
      'módulo da força = 100N\n'
      'gravidade = 10 m/s²\n'
      'ângulo da força com a direção horizontal = 38 graus.\n\n'
      'Sabendo que o atrito está sendo desprezado, responda: \n')

while True:
    try:
        q2 = float(input('a) Qual o valor, em Newtons (N), da componente x da Força? (Duas casas decimais. Considere as regras de arredondamento e \n '
                         'utilize ponto, não vírgula.) '))
        if q2 == Fx:
            print_verde('Parabéns! Resposta correta.')
            print('---------------------------------------------------------------------------------------------')
            break
        else:
            while q2 != Fx:
                print_vermelho('Resposta incorreta. ')
                q2 = float(input('Tente novamente. Dica: aqui devemos encontrar a componente horizontal da força. \nPara isso, devemos utilizar as relações trigonométricas.'))
                if q2 == Fx:
                    print_verde('Parabéns! Resposta correta.')
                    print('--------------------------------------------------------------------------------'
                          '-------------')
                    break
            break
    except ValueError:
        print('Digite em números, apenas.')


while True:
    try:
        q3 = float(input('b) Qual o valor, em Newtons (N), da componente y da Força? (Duas casas decimais. Considere as regras de arredondamento e \n '
                         'utilize ponto, não vírgula.) '))
        if q3 == Fy:
            print_verde('Parabéns! Resposta correta.')
            print('---------------------------------------------------------------------------------------------')
            break
        else:
            while True:
                print_vermelho('Resposta incorreta. ')
                q3 = float(input('Tente novamente. Dica: utilize as relações trigonométricas para encontrar a componente vertical da força.'))
                if q3 == Fy:
                    print_verde('Parabéns! Resposta correta.')
                    print('-------------------------------------------------------------------------------------'
                          '--------')
                    break
            break
    except ValueError:
        print('Digite em números, apenas.')

while True:
    try:
        q4 = float(input('c) Determine o módulo, em Newtons (N), da força peso que age sobre bloco. '))
        if q4 == P:
            print_verde('Parabéns! Resposta correta.')
            print('---------------------------------------------------------------------------------------------')
            break
        else:
            while True:
                print_vermelho('Resposta incorreta. ')
                q4 = float(input('Tente novamente. Utilize os dados fornecidos na introdução da questão 1 para encontrar a força peso.'))
                if q4 == P:
                    print_verde('Parabéns! Resposta correta.')
                    print('-------------------------------------------------------------------------------------'
                          '--------')
                    break
            break
    except ValueError:
        print('Digite em números, apenas.')

while True:
    try:
        q5 = float(input('d) Determine o módulo, em Newtons (N), da força normal que age sobre o bloco. (Duas casas decimais. Considere as regras de arredondamento e \n '
                         'utilize ponto, não vírgula.)  '))
        if q5 == Normal:
            print_verde('Parabéns! Resposta correta.')
            print('---------------------------------------------------------------------------------------------')
            break
        else:
            while True:
                print_vermelho('Resposta incorreta. ')
                q5 = float(input('Tente novamente. Dica: a força normal tem o mesmo valor, com sentido oposto, da resultante das forças (excluindo a propria normal) que agem na direção \nperpendicular à superfície de apoio. '
                                 'Neste caso, devemos encontrar a resultante entre'
                                 ' o Peso e a componente y da Força.'))
                if q5 == Normal:
                    print_verde('Parabéns! Resposta correta.')
                    print('-------------------------------------------------------------------------------------'
                          '--------')
                    break
            break
    except ValueError:
        print('Digite em números, apenas.')

while True:
    try:
        q6 = float(input('e) Determine a aceleração do bloco, sabendo que ele se move apenas na horizontal. \n'
                         '1) 3.152 m/s²\n'
                         '2) 2.460 m/s²\n'
                         '3) 4 m/s²\n'
                         '4) NDA\n'
                         'Indique a alternativa correta (digite 1, 2, 3 ou 4).\n'))
        if q6 == 1:
            print_verde('Parabéns! Resposta correta.')
            print('---------------------------------------------------------------------------------------------')
            break
        else:
            while q6 != 1:
                print_vermelho('Resposta incorreta. ')
                q6 = float(input('Tente novamente. Lembre-se que Força = massa * aceleração. Dica: como o bloco se '
                                 'move apenas na horizontal, devemos considerar somente as forças nessa direção.'))
                if q6 == 1:
                    print_verde('Parabéns! Resposta correta.')
                    print('-------------------------------------------------------------------------------------'
                          '--------')
                    break
            break
    except ValueError:
        print('Digite em números, apenas.')

print_verde('Você concluiu a trilha de exercícios.')
```

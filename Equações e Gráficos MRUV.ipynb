```python
import numpy as np
import matplotlib.pyplot as plt


print('      EQUAÇÕES E GRÁFICOS MRUV      ')
print('------------------------------------')
xo = float(input('Posição inicial (m): '))
vo = float(input('Velocidade inicial (m/s): '))
a = float(input('Aceleração (m/s^2): '))
t = float(input('Tempo (s): '))
h = t
print('------------------------------------')


def x(t):
    return xo+vo+(a/2)*t**2


plt.xlim(0, t)
plt.ylim(0, x(t))
t = np.linspace(0, t, 200)
plt.plot(t, x(t), color='red', linewidth=1)
plt.xlabel('Tempo (s)')
plt.ylabel('Posição (m)')
plt.title('MRUV: POSIÇÃO x TEMPO')
plt.grid(True)
plt.savefig('posição.png')
plt.show()


def v(t):
    return vo+a*t


plt.xlim(0, h)
plt.ylim(0, v(h))
t = np.linspace(0, h, 200)
plt.plot(t, v(t), color='blue', linewidth=1)
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade (m/s)')
plt.title('MRUV: VELOCIDADE x TEMPO')
plt.grid(True)
plt.savefig('velocidade.png')
plt.show()


def ace(t):
    return a+t*0


plt.xlim(0, h)
plt.ylim(0, a*2)
t = np.linspace(0, h, 200)
plt.plot(t, ace(t), color='green', linewidth=1)
plt.xlabel('Tempo (s)')
plt.ylabel('Aceleração (m/s²)')
plt.title('MRUV: ACELERAÇÃO x TEMPO')
plt.grid(True)
plt.savefig('aceleração.png')
plt.show()
```

```python
from vpython import *


def print_verde(texto):
    print('\033[32m{}\033[0m'.format(texto))


def print_vermelho(texto):
    print('\033[31m{}\033[0m'.format(texto))


def run_simulation():
    # Remover esferas e setas existentes
    for obj in scene.objects:
        obj.visible = False

    # Coordenadas iniciais das esferas
    x1, y1, z1 = 2, 0, 0
    x2, y2, z2 = 8, 0, 0
    x3, y3, z3 = 2, 3, 0
    x4, y4, z4 = 8, 3, 0
    x5, y5, z5 = 2, 6, 0
    x6, y6, z6 = 8, 6, 0

    # Vetores de velocidade das esferas
    v1 = vector(2, 0, 0)  # Velocidade inicial da esfera azul
    v2 = vector(0, 0, 0)  # Velocidade inicial da esfera vermelha
    v3 = vector(1, 0, 0)  # Velocidade inicial da esfera verde
    v4 = vector(-1, 0, 0)  # Velocidade inicial da esfera amarela
    v5 = vector(2, 0, 0)  # Velocidade inicial da esfera ciano
    v6 = vector(0, 0, 0)  # Velocidade inicial da esfera magenta

    # Criando as esferas
    sphere1 = sphere(pos=vector(x1, y1, z1), radius=0.5, color=color.blue)
    sphere2 = sphere(pos=vector(x2, y2, z2), radius=0.5, color=color.red)
    sphere3 = sphere(pos=vector(x3, y3, z3), radius=0.5, color=color.green)
    sphere4 = sphere(pos=vector(x4, y4, z4), radius=0.5, color=color.yellow)
    sphere5 = sphere(pos=vector(x5, y5, z5), radius=0.5, color=color.cyan)
    sphere6 = sphere(pos=vector(x6, y6, z6), radius=0.5, color=color.magenta)

    # Adicionando setas de velocidade acima das esferas
    arrow1 = arrow(pos=sphere1.pos + vector(0, 0.75, 0), axis=1.5 * v1, color=color.blue)
    arrow2 = arrow(pos=sphere2.pos + vector(0, 0.75, 0), axis=1.5 * v2, color=color.red)
    arrow3 = arrow(pos=sphere3.pos + vector(0, 0.75, 0), axis=1.5 * v3, color=color.green)
    arrow4 = arrow(pos=sphere4.pos + vector(0, 0.75, 0), axis=1.5 * v4, color=color.yellow)
    arrow5 = arrow(pos=sphere5.pos + vector(0, 0.75, 0), axis=1.5 * v5, color=color.cyan)
    arrow6 = arrow(pos=sphere6.pos + vector(0, 0.75, 0), axis=1.5 * v6, color=color.magenta)

    # Parâmetros da cena
    scene.range = 10
    scene.autoscale = False

    # Definindo o tempo e o passo de tempo
    t = 0
    dt = 0.01

    # Simulação da colisão
    while t < 5:
        rate(100)

        # Atualizando as posições
        sphere1.pos += v1 * dt
        sphere2.pos += v2 * dt
        sphere3.pos += v3 * dt
        sphere4.pos += v4 * dt
        sphere5.pos += v5 * dt
        sphere6.pos += v6 * dt

        # Verificando colisões
        if (sphere2.pos.x - sphere1.pos.x) * (v2.x - v1.x) < 0 and mag(sphere1.pos - sphere2.pos) < 1:
            v2 = v1
            v1 = vector(0, 0, 0)
            arrow1.axis = 1.5 * v1
            arrow2.axis = 1.5 * v2

        if mag(sphere3.pos - sphere4.pos) < 1:
            v4 = -v4
            v3 = -v3
            arrow3.axis = 1.5 * v3
            arrow4.axis = 1.5 * v4

        if mag(sphere5.pos - sphere6.pos) < 1:
            v6 = vector(1, 0, 0)
            v5 = vector(1, 0, 0)
            arrow5.axis = 1.5 * v5
            arrow6.axis = 1.5 * v6

        t += dt


# Inicializar a simulação
run_simulation()

# Solicitar entrada do usuário
while True:
    user_input = input('Deseja ver a simulação novamente? (s/n): ').lower()
    if user_input == 's':
        run_simulation()
    elif user_input == 'n':
        break
    else:
        print('Por favor, digite "s" para sim ou "n" para não.')

print('---------------------------------------------------------------------------------------------')
print('                                COLISÕES                                ')
print('---------------------------------------------------------------------------------------------')
print('Em física, colisões referem-se ao estudo das interações entre objetos em movimento.')
print('A simulação anterior representa os três tipos principais tipos de colisão a serem estudados:\n'
      'Situação 1 - as esferas seguem "grudadas" após a colisão.\n'
      'Situação 2 - as esferas seguem direções opostas após a colisão.\n'
      'Situação 3 - uma das esferas transfere totalmente sua velocidade para a outra após a colisão.\n')
print('---------------------------------------------------------------------------------------------')
print('                                       QUESTÃO 1                                             ')
print('---------------------------------------------------------------------------------------------')
print('Considere que duas esferas tenham a massa M. Uma delas está em repouso e a outra se move '
      'em direção à colisão com velocidade V.\n Após a colisão, qual das situações ocorre?\n')

while True:
    try:
        q1 = int(input('1 - Situação 1\n'
                       '2 - Situação 2\n'
                       '3 - Situação 3\n\n'
                       'Digite o número da alternativa correta.\n'))
        if q1 in [1, 2, 3]:
            break
        else:
            print('Por favor, digite um número entre 1 e 3.')
    except ValueError:
        print_vermelho('Por favor, digite um número inteiro válido.')

if q1 == 3:
    print_verde('Parabéns! Resposta correta.')
else:
    while True:
        try:
            if q1 == 1:
                print_vermelho('Resposta incorreta.')
                q1 = int(input(
                    'Dica: como as massas são iguais, ocorre uma colisão elástica, \nonde a Energia Cinética inicial é conservada após o choque.\n'
                    'Isso não é observado na situação 1.'))
            elif q1 == 2:
                print_vermelho('Resposta incorreta.')
                q1 = int(input(
                    'Como ambas as massas são iguais, ocorre uma conservação da Energia Cinética, o que não é observado na situação 2.\n'))
            elif q1 == 3:
                print_verde('Resposta correta.')
                break
            else:
                q1 = int(input('Por favor, digite um número entre 1 e 3: '))
        except ValueError:
            print_vermelho('Por favor, digite um número inteiro válido.')

print('---------------------------------------------------------------------------------------------')
print('                                       QUESTÃO 2                                             ')
print('---------------------------------------------------------------------------------------------')
print(
    'Considere duas esferas: azul e rosa. Elas se colidem e ocorre a situação 1, onde ambas seguem "grudadas" após o choque.\n'
    'Dados:\n\n'
    'massa azul: 0.299 kg\n'
    'velocidade inicial azul: 0.6 m/s\n'
    'massa rosa: 0.205 kg\n'
    'velocidade inicial rosa: 0 m/s\n'
    'velocidade final azul e rosa: 0.283 m/s\n')

ma = 0.299
via = 0.6
mr = 0.205
vir = 0
vfa = 0.283
vfr = vfa

Ki_azul = (ma * (via ** 2)) / 2
Kf_azul = (ma * (vfa ** 2)) / 2
Ki_rosa = (mr * (vir ** 2)) / 2
Kf_rosa = (mr * (vfr ** 2)) / 2

Ki = round((Ki_azul + Ki_rosa), 3)
Kf = round((Kf_azul + Kf_rosa), 3)

# print(Kf_rosa, Kf_azul)

# print(Ki, Kf)

while True:
    try:
        q1 = int(input('a) Neste caso, ocorre conservação de energia cinética?\n'
                       '1 - Sim\n'
                       '2 - Não\n'))
        if q1 in [1, 2]:
            break
        else:
            print('Por favor, digite um número entre 1 e 2.')
    except ValueError:
        print_vermelho('Por favor, digite um número inteiro válido.')

if q1 == 2:
    print_verde('Parabéns! Resposta correta.')
else:
    while True:
        try:
            if q1 == 1:
                print_vermelho('Resposta incorreta.')
                q1 = int(input('Este tipo de colisão é chamado de "colisão perfeitamente inelástica", onde não há conservação\n'
                    'da Energia Cinética. Tente novamente.\n'))
            elif q1 == 2:
                print_verde('Resposta correta.')
                break
            else:
                q1 = int(input('Por favor, digite um número entre 1 e 2: '))
        except ValueError:
            print_vermelho('Por favor, digite um número inteiro válido.')



while True:
    try:
        q3 = float(input('b) Calcule a Energia Cinética inicial total do sistema. Utilize dois algarismos significativos.'))
        if q3 == Ki:
            print_verde('Resposta correta.')
            break
        else:
            print_vermelho('Resposta incorreta.')
            print('Encontre a Energia Cinética inicial para ambas as esferas e some as duas. Lembre-se das \n'
            'regras de arredondamento.')
    except ValueError:
        print_vermelho('Por favor, digite um número válido, com ponto como separador decimal.')

while True:
    try:
        q4 = float(input('c) Calcule a Energia Cinética final total do sistema. Utilize dois algarismos significativos.'))
        if q4 == Kf:
            print_verde('Resposta correta.')
            break
        else:
            print_vermelho('Resposta incorreta.')
            print('Encontre a Energia Cinética final para ambas as esferas e some as duas. Lembre-se das \n'
            'regras de arredondamento.')
    except ValueError:
        print_vermelho('Por favor, digite um número válido, com ponto como separador decimal.')

print('---------------------------------------------------------------------------------------------')
print('                                       QUESTÃO 3                                             ')
print('---------------------------------------------------------------------------------------------')
print(
    'Considere duas esferas, 1 e 2. Elas se colidem e ocorre a situação 2, onde elas se movem\n'
    'em direções opostas após a colisão.\n'
    'Dados:\n\n'
    'massa 1: 0.200 kg\n'
    'velocidade inicial 1: 0.916 m/s\n'
    'velocidade final 1: -0,2175 m/s\n'
    'massa 2: 0.305 kg\n'
    'velocidade inicial 2: 0 m/s\n'
    'velocidade final 2: 0.611 m/s\n')

m1 = 0.2
v1_inicial = 0.916
v1_final = -0.2175
m2 = 0.305
v2_inicial = 0
v2_final = 0.611

p1_inicial = v1_inicial*m1
p2_inicial = v2_inicial*m2
p1_final = v1_final*v1_final
p2_final = v2_final*v2_final

p_inicial = round(p1_inicial + p2_inicial, 2)
p_final = round(p1_final + p2_final, 2)

print(p_inicial, p_final)

while True:
    try:
        q4 = float(input('Calcule o momento inicial total do sistema. Utilize dois algarismos significativos.'))
        if q4 == p_inicial:
            print_verde('Resposta correta.')
            break
        else:
            print_vermelho('Resposta incorreta.')
            print('Encontre o momento inicial para ambas as esferas e some os dois. Lembre-se das \n'
            'regras de arredondamento.')
    except ValueError:
        print_vermelho('Por favor, digite um número válido, com ponto como separador decimal.')

while True:
    try:
        q4 = float(input('Calcule o momento final total do sistema. Utilize dois algarismos significativos.'))
        if q4 == p_final:
            print_verde('Resposta correta.')
            break
        else:
            print_vermelho('Resposta incorreta.')
            print('Encontre o momento final para ambas as esferas e some os dois. Lembre-se das \n'
            'regras de arredondamento.')
    except ValueError:
        print_vermelho('Por favor, digite um número válido, com ponto como separador decimal.')

while True:
    try:
        q1 = int(input('Neste caso, é uma colisão elástica??\n'
                       '1 - Sim\n'
                       '2 - Não\n'))
        if q1 in [1, 2]:
            break
        else:
            print('Por favor, digite um número entre 1 e 2.')
    except ValueError:
        print_vermelho('Por favor, digite um número inteiro válido.')

if q1 == 2:
    print_verde('Parabéns! Resposta correta.')
else:
    while True:
        try:
            if q1 == 1:
                print_vermelho('Resposta incorreta.')
                q1 = int(input('Na colisão elástica ocorre conservação da energia cinética e do momento linear.\n'))
            elif q1 == 2:
                print_verde('Resposta correta.')
                break
            else:
                q1 = int(input('Por favor, digite um número entre 1 e 2: '))
        except ValueError:
            print_vermelho('Por favor, digite um número inteiro válido.')

print('Parabéns! Você concluiu a trilha de exercícios.')
```

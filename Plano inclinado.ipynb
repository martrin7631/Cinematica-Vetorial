```python
# Centro Universitário FEI

# Criado por Marcelo Santos Trindade em 12/11/2023

# "Elaboração de programas em Python como apoio didático na disciplina de Física I para estudantes ingressantes".

import matplotlib.pyplot as plt
import numpy as np


# Função para imprimir o texto em verde ou vermelho
def print_verde(texto):
    print('\033[32m{}\033[0m'.format(texto))


def print_vermelho(texto):
    print('\033[31m{}\033[0m'.format(texto))


# Coordenadas dos vértices do triângulo
triangle_x = [-1.5, 3, -1.5]
triangle_y = [0, 0, 6]

# Coordenadas dos vértices do quadrado
square_x = [1.5, 2.5, 1, 0, 1.5]
square_y = [2, 3, 5, 4, 2]

# linha
line_x = [-1.5, -1.5]
line_y = [0, 6]

# Plota o triângulo e o quadrado
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.plot(triangle_x, triangle_y, 'b-', linewidth=2)  # Linha azul para representar o triângulo
ax.plot(square_x, square_y, 'r-', linewidth=2)  # Linha vermelha para representar o quadrado
ax.plot(line_x, line_y, 'b-', linewidth=2)
ax.fill(triangle_x, triangle_y, 'lightblue')  # Preenche o triângulo com a cor lightblue
ax.fill(square_x, square_y, 'salmon')  # Preenche o quadrado com a cor salmon
ax.set_aspect('equal')
plt.xlim(-4, 4)  # Define os limites do eixo x
plt.ylim(-4, 8)  # Define os limites do eixo y
plt.title('Plano inclinado')

print('                                PLANO INCLINADO                                ')
print('---------------------------------------------------------------------------------------------')
print('O plano inclinado é uma superfície plana, elevada e inclinada, por exemplo, uma rampa. Na física, \n'
      'Estudamos o movimento dos objetos bem como a aceleração e as forças atuantes que ocorrem \n'
      'num plano inclinado.\n'
      ''
      'Existem 2 tipos de forças que atuam nesse sistema sem atrito: a força normal, que faz 90º em \n'
      'relação ao plano, e a força peso (força vertical para baixo). Note que elas possuem direções \n'
      'e sentidos diferentes.')
print('---------------------------------------------------------------------------------------------')

plt.show()

m = 5
a = 10
P = m*a
angle_rad = np.radians(60)
sen60 = round(np.sin(angle_rad), 2)
cos60 = round(np.cos(angle_rad), 2)
Px = round(P*sen60, 2)
Py = round(P*cos60, 2)
abloco = (Px-10)/m

#print(m, a, P,sen60,cos60,Px,Py,abloco)
#print("peso =", P)


print('---------------------------------------------------------------------------------------------')
print('                                       QUESTÃO 1                                             ')
print('---------------------------------------------------------------------------------------------')
print('A figura mostra um bloco sobre um plano inclinado (não está em escala).\n\n'
      'Considere:\n'
      'massa do bloco = 5kg\n'
      'aceleração da gravidade = 10 m/s²\n'
      'ângulo de inclinação do plano = 60 graus.\n\n'
      'Encontre os valores de seno e cosseno com a calculadora e utilize-os com duas casas decimais \npara os cálculos (lembre-se de aplicar as regras de arredondamento).\n\n'
      'Sabendo que o atrito está sendo desprezado, responda: \n')

while True:
    try:
        q1 = float(input('a) Determine, em Newtons (N), a força peso que age sobre o bloco.'))
        if q1 == P:
            print_verde('Parabéns! Resposta correta.')
            print('---------------------------------------------------------------------------------------------')
            break
        else:
            while q1 != P:
                print_vermelho('Resposta incorreta. ')
                q1 = float(input('Tente novamente. Lembre-se que o peso é sempre vertical e depende apenas da massa '
                                 'e da aceleração da gravidade.) '))
                if q1 == P:
                    print_verde('Parabéns! Resposta correta.')
                    print('--------------------------------------------------------------------------------'
                          '-------------')
                    break
            break
    except ValueError:
        print_vermelho('Digite em números, apenas.')

while True:
    try:
        q2 = float(input('b) Determine, em Newtons (N), a componente x da força peso. (Duas casas decimais, '
                         'respeitando as regras de arredondamento. Utilize ponto, não vírgula)'))
        if q2 == Px:
            print_verde('Parabéns! Resposta correta.')
            print('---------------------------------------------------------------------------------------------')
            break
        else:
            while q2 != Px:
                print_vermelho('Resposta incorreta. ')
                q2 = float(input('Tente novamente. Dica: Neste caso, você deve calcular a componente da força peso '
                'paralela ao plano inclidado. \nUtilize o valor de seno ou cosseno com duas casas decimais para o cálculo, e lembre-se das regras de arredondamento!'))
                if q2 == Px:
                    print_verde('Parabéns! Resposta correta.')
                    print('--------------------------------------------------------------------------------'
                          '-------------')
                    break
            break
    except ValueError:
        print_vermelho('Digite em números, apenas.')

while True:
    try:
        q3 = float(input('c) Determine, em Newtons (N), a força normal. (Duas casas decimais, '
                         'respeitando as regras de arredondamento. Utilize ponto, não vírgula)'))
        if q3 == Py:
            print_verde('Parabéns! Resposta correta.')
            print('---------------------------------------------------------------------------------------------')
            break
        else:
            while q3 != Py:
                print_vermelho('Resposta incorreta. ')
                q3 = float(input('Tente novamente. Dica: Neste caso, força normal é igual à componente y do peso, que '
                'é perpendicular ao plano inclinado. \nUtilize o valor de seno ou cosseno com duas casas decimais para o cálculo, e lembre-se das regras de arredondamento!'))
                if q3 == Py:
                    print_verde('Parabéns! Resposta correta.')
                    print('--------------------------------------------------------------------------------'
                          '-------------')
                    break
            break
    except ValueError:
        print_vermelho('Digite em números, apenas.')

while True:
    try:
        q4 = float(input('d) Considere que o bloco sofre ação de uma força de 10N, contrária ao sentido do'
                         ' deslizamento. \nDetermine a aceleração (m/s^2) do bloco nessa situação. (Duas casas decimais, '
                         'respeitando as regras de arredondamento. Utilize ponto, não vírgula)'))
        if q4 == abloco:
            print_verde('Parabéns! Resposta correta.')
            print('---------------------------------------------------------------------------------------------')
            break
        else:
            while q4 != abloco:
                print_vermelho('Resposta incorreta. ')
                q4 = float(input('Tente novamente. Dica: A força resultante (Fr = m*a) é a resultante de todas as forças \n'
                'que atuam de maneira paralela ao plano.'))
                if q4 == abloco:
                    print_verde('Parabéns! Resposta correta.')
                    print('--------------------------------------------------------------------------------'
                          '-------------')
                    break
            break
    except ValueError:
        print_vermelho('Digite em números, apenas.')

print_verde('Você concluiu a trilha de exercícios.')
```

```python
# Centro Universitário FEI

# Criado por Marcelo Santos Trindade em 17/02/2023

# "Elaboração de programas em Python como apoio didático na disciplina de Física I para estudantes ingressantes".

# Programa desenvolvido como parte de um projeto de Iniciação Didática com objetivo de estudar Python como ferramenta
# auxiliar para o ensino de Física I no Centro Universitário FEI.

import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import math


# Função para imprimir o texto em verde ou vermelho
def print_verde(texto):
    print('\033[32m{}\033[0m'.format(texto))


def print_vermelho(texto):
    print('\033[31m{}\033[0m'.format(texto))


# Cria uma figura e os eixos de coordenadas X e Y associados
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect('equal')  # torna a escala dos eixos X e Y iguais, deixando o círculo proporcional
plt.grid(True, zorder=1)
plt.title('Circunferência Trigonométrica', fontsize=18)
plt.xticks([i/2 for i in range(-5, 5)], fontsize=10)
plt.yticks([i/2 for i in range(-5, 5)], fontsize=10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)

# Cria uma circunferência com centro (0,0) e raio 1
circle = patches.Circle((0, 0), radius=1, edgecolor='red', fill=False, alpha=1, zorder=2)
ax.add_patch(circle)

# Adiciona linhas e anotações para os ângulos notáveis
angles = [0, np.pi/6, np.pi/4, np.pi/3, np.pi/2, np.pi, 3*np.pi/2]
angle_names = [r'$0$', r'$\frac{\pi}{6}$', r'$\frac{\pi}{4}$', r'$\frac{\pi}{3}$', r'$\frac{\pi}{2}$', r'$\pi$',
               r'$\frac{3\pi}{2}$', r'$2\pi$']
for i, angle in enumerate(angles):
    x = np.cos(angle)
    y = np.sin(angle)
    ax.plot([0, x], [0, y], color='black', lw=1)
    ax.text(x*1.1, y*1.1, angle_names[i], ha='center', va='center', fontsize=20)

print('                                CIRCUNFERÊNCIA TRIGONOMÉTRICA                                ')
print('---------------------------------------------------------------------------------------------')
print('A Circunferência Trigonométrica, como mostra a figura, é uma circunferência de raio 1,\n'
      'centrada na origem (0,0).\n'
      'Em cada ponto dessa circunferência, a coordenada horizontal (abscissa) representa o cosseno\n'
      'do ângulo, enquanto a coordenada vertical (ordenada) representa o seno do ângulo.\n')
print('---------------------------------------------------------------------------------------------')

plt.show()

print('---------------------------------------------------------------------------------------------')
print('                                       QUESTÃO 1                                             ')
print('---------------------------------------------------------------------------------------------')
print('Os arcos correspondentes ao arco de pi/3, nos demais quadrantes, valem, em graus:\n'
      '1) 150, 210, 330\n'
      '2) 120, 240, 300\n'
      '3) 135, 225, 315\n'
      '4) 90, 180, 270\n')

while True:
    try:
        q1 = int(input('Digite o número da alternativa correta (1, 2, 3 ou 4): '))
        if q1 in [1, 2, 3, 4]:
            break
        else:
            print_vermelho('Por favor, digite um número entre 1 e 4.')
    except ValueError:
        print_vermelho('Por favor, digite um número inteiro válido.')

if q1 == 2:
    print_verde('Parabéns! Resposta correta.')
else:
    while True:
        if q1 == 1:
            print_vermelho('Resposta incorreta.')
            q1 = int(input('Os ângulos da alternativa 1 correspondem ao arco de 30°. Tente '
                           'novamente: '))
        elif q1 == 3:
            print_vermelho('Resposta incorreta.')
            q1 = int(input('Os ângulos da alternativa 3 correspondem ao arco de 45°. Tente '
                           'novamente: '))
        elif q1 == 4:
            print_vermelho('Resposta incorreta.')
            q1 = int(input('Os ângulos da alternativa 4 correspondem ao arco de 90°. Tente '
                           'novamente: '))
        elif q1 == 2:
            print_verde('Resposta correta.')
            break
        else:
            q1 = int(input('Resposta incorreta. Por favor, digite um número entre 1 e 4: '))

print('---------------------------------------------------------------------------------------------\n')

print('---------------------------------------------------------------------------------------------')
print('                                       QUESTÃO 2                                             ')
print('---------------------------------------------------------------------------------------------')
print('Os arcos correspondentes ao arco de 330°, nos demais quadrantes, valem: \n'
      '1) 50, 180, 210\n'
      '2) 60, 150, 300\n'
      '3) 30, 150, 210\n'
      '4) 40, 160, 220\n'
      '5) 120, 200, 340\n')

try:
    q1 = float(input('Digite o número da alternativa correta (1, 2, 3, 4 ou 5): '))
except:
    q1 = float(input('Digite em números, apenas (utilize ponto, não virgula): '))
if q1 == 3:
    print_verde('Parabéns! Resposta correta.')
while q1 != 3:
    print_vermelho('Resposta incorreta.')
    q1 = float(input('Tente novamente: '))
    if q1 == 3:
        print_verde('Parabéns! Resposta correta.')
        break
print('---------------------------------------------------------------------------------------------')

print('---------------------------------------------------------------------------------------------')
print('                                       QUESTÃO 3                                             ')
print('---------------------------------------------------------------------------------------------')
print('Analise o seguinte triângulo (as unidades são arbitrárias): ')

# Lados do triângulo
a = 30
b = 40
c = math.sqrt(a**2 + b**2)

# Coordenadas do triângulo
x = [0, a, a, 0]
y = [0, 0, b, 0]

# Desenha o triângulo
plt.plot(x, y)

plt.grid(True, zorder=-1)
plt.ylim(-10, 40)

# Adiciona o valor do cateto horizontal
plt.text(a/2, -2.75, f'{a}', ha='center')

# Adiciona o valor do cateto vertical
plt.text(a+0.5, b/2, f'{b}', va='center')


plt.show()


try:
    q2 = float(input('a) Qual o valor da hipotenusa? '))
except:
    q2 = float(input('Digite em números, apenas: '))

if q2 == c:
    print_verde('Parabéns! Resposta correta.')
    print('---------------------------------------------------------------------------------------------')
while q2 != c:
    print_vermelho('Resposta incorreta.')
    q2 = float(input('Tente novamente. (Dica: hip^2 = cateto1^2 + cateto2^2): '))
    if q2 == c:
        print_verde('Parabéns! Resposta correta.')
        print('---------------------------------------------------------------------------------------------')

seno = b/c

try:
    q3 = float(input('b) Qual o valor do seno do ângulo formado entre o cateto horizontal e a hipotenusa? '
                     '(Uma casa decimal. Utilize ponto, não vírgula.) '))
except:
    q3 = float(input('Digite em números, apenas (utilize ponto, não virgula): '))

if q3 == seno:
    print_verde('Parabéns! Resposta correta.')
    print('---------------------------------------------------------------------------------------------')
while q3 != seno:
    print_vermelho('Resposta incorreta.')
    q3 = float(input('Tente novamente. (Dica: seno = cateto oposto / hipotenusa): '))
    if q3 == seno:
        print('---------------------------------------------------------------------------------------------')
        print_verde('Parabéns! Resposta correta.')
        print('---------------------------------------------------------------------------------------------\n')

print('---------------------------------------------------------------------------------------------')
print('                                       QUESTÃO 4                                             ')
print('---------------------------------------------------------------------------------------------')
print('Analise o seguinte triângulo (as unidades são arbitrárias):\n')

# Lados do triângulo
hipotenusa = 100
cateto1 = math.cos(math.radians(50)) * hipotenusa
cateto2 = math.sin(math.radians(50)) * hipotenusa

# Coordenadas do triângulo
x = [0, cateto1, 0, 0]
y = [0, 0, cateto2, 0]

# Desenha o triângulo
plt.plot(x, y)

plt.grid(True, zorder=-1)
plt.ylim(-10, 110)
plt.xlim(-10, 110)

# Adiciona o valor da hipotenusa
plt.text(cateto1/2, cateto2/2, f'{hipotenusa}', ha='center', va='center')

# Calcula as coordenadas do ângulo alfa
angulo_alfa = math.radians(50)
x_angulo_alfa = cateto1 * 0.8
y_angulo_alfa = cateto2 * 0.1

# Adiciona o ângulo alfa
plt.text(x_angulo_alfa, y_angulo_alfa, r'$\alpha = 50^\circ$', ha='center')

plt.show()

print('a) Em relação ao ângulo alfa, calcule o valor do cateto oposto.\n'
      'Dados: sen50 = 0,76  cos50 = 0,64\n')

while True:
    try:
        q1 = float(input('Digite o valor do cateto oposto: '))
        break
    except ValueError:
        print_vermelho('Por favor, digite um número inteiro válido.')

if q1 == 76:
    print_verde('Parabéns! Resposta correta.')
else:
    while True:
        if q1 != 76:
            print_vermelho('Resposta incorreta.')
            q1 = float(input('Lembre-se: sen = (CO/HIP): '))
        elif q1 == 76:
            print_verde('Parabéns! Resposta correta.')
            break
        else:
            print_vermelho('Resposta incorreta')
            q1 = float(input('Por favor, digite um número válido: '))


print('\nb) Calcule, agora, o valor do cateto adjacente: \n')

while True:
    try:
        q2 = float(input('Digite o valor do cateto adjacente: '))
        break
    except ValueError:
        print_vermelho('Por favor, digite um número inteiro válido. ')

if q2 == 64:
    print_verde('Parabéns! Resposta correta.')
else:
    while True:
        if q2 != 64:
            print_vermelho('Resposta incorreta.')
            q2 = float(input('Lembre-se: cos = (CA/HIP): '))
        elif q2 == 64:
            print_verde('Parabéns! Resposta correta.')
            break
        else:
            print_vermelho('Resposta incorreta')
            q2 = float(input('Por favor, digite um número válido: '))

print('\nc) Com os valores obtidos, calcule o valor da tangente do ângulo alfa.\n'
      '1) 1,1875\n'
      '2) 0,8421\n'
      '3) 1\n'
      '4) NDA\n')

while True:
    try:
        q1 = int(input('Digite o número da alternativa correta (1, 2, 3 ou 4): '))
        if q1 in [1, 2, 3, 4]:
            break
        else:
            print_vermelho('Por favor, digite um número entre 1 e 4.')
    except ValueError:
        print_vermelho('Por favor, digite um número inteiro válido.')

if q1 == 1:
    print_verde('Parabéns! Resposta correta.')
else:
    while True:
        if q1 == 2:
            print_vermelho('Resposta incorreta')
            q1 = int(input('Lembre-se: tg = CO/CA. Tente novamente: '))
        elif q1 == 3:
            print_vermelho('Resposta incorreta')
            q1 = int(input('Lembre-se: tg = CO/CA. Tente novamente: '))
        elif q1 == 4:
            print_vermelho('Resposta incorreta')
            q1 = int(input('Lembre-se: tg = CO/CA. Tente novamente: '))
        elif q1 == 1:
            print_verde('Resposta correta.')
            break
        else:
            print_vermelho('Resposta incorreta')
            q1 = int(input('Por favor, digite um número entre 1 e 4: '))

print('\n\n\nParabéns! Você concluiu a trilha de exercícios.')
```
